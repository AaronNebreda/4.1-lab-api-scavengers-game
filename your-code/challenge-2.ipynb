```python
count = 0 
lista_commits = []

for i in range(len(res.json())):
    fecha = res.json()[i]['commit']['author']['date']
    if fecha > date_ini and fecha < date_fin:
        count += 1
        lista_commits.append(res.json()[i])
        
        
        
    
```
